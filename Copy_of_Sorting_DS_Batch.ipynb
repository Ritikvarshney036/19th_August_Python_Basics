```python
#bubble sorting

def bubbleSort(arr):
  for i in range(len(arr)-1, 0 , -1):
    for j in range(i):
      if(arr[j]>arr[j+1]):
        arr[j],arr[j+1] = arr[j+1],arr[j]
```

# New Section

```python
arr = [5,4,3,2,1]
bubbleSort(arr)
print(arr)
```

```python
def bubbleSortOptimized(arr):
  for i in range(len(arr)-1, 0 , -1):
    isSorted = True
    for j in range(i):
      if(arr[j]>arr[j+1]):
        isSorted = False
        arr[j],arr[j+1] = arr[j+1],arr[j]

    if isSorted :
      print("Array is already sorted")
      break
```

```python
arr = [1,2,3,5,4]
bubbleSortOptimized(arr)
print(arr)
```

```python
# Given an array, check if the array is sorted ?

def isSorted(arr):
  isSorted = True
  for i in range(len(arr)-1):
    if arr[i]>arr[i+1]:
      isSorted = False
      break
  if isSorted :
    print("Array is sorted")
  else :
    print("Array is not sorted")
```

```python
#Selection Sorting
def selectionSort(arr):
  for i in range(len(arr)):
    min =i
    for j in range(i+1,len(arr)):
      if arr[j] <arr[min]:
        min =j
    #Find the smallest values
    arr[i],arr[min] = arr[min],arr[i]
```

```python
def insertion_sort(arr):
  for i in range(1,len(arr)):
    v = arr[i]
    j=i
    while(j>=1 and arr[j-1]>v):
      arr[j]=arr[j-1]
      j-=1
    arr[j]=v
```

```python
def search(arr, num):
  for i in range(len(arr)):
    if arr[i]==num:
      return i
  return -1

```

```python
print(search([4,5,1,5,9],5))
```

```python
def binarySearch(arr, num):
  left = 0
  right = len(arr) -1


  while(left <=right):
    mid = int(left + (right-left)/2)
    if arr[mid] == num:
      return mid
    elif arr[mid] > num :
      right = mid-1
    else :
      left = mid+1
  return -1


```

```python
def printRec(arr, sI):
  #Base condition
  if sI >= len(arr):
    return

  #Logic
  print(arr[sI])

  #Recursive call
  printRec(arr,sI+1)




printRec([1,2,3,4],0)
```

```python
def find_min(arr, sI):
  #Base condition
  if sI >=len(arr):
    return 99999999

  return min(arr[sI], find_min(arr,sI+1))


print(find_min([3,2,5],0))
```

```python
def moves(n, s ,d , a):
  # Base condition
  if n == 1 :
    return 1

  # move n-1 s->a , then move 1 biggest coin to s->d and then n-1 coin a-d
  return moves(n-1,s,a,d) + 1 + moves(n-1,a,d,s)

print(moves(1,'s','d','a'))
print(moves(2,'s','d','a'))
print(moves(3,'s','d','a'))
print(moves(200,'s','d','a'))
```
